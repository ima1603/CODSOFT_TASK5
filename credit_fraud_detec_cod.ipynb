# feature selection

```python
X=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import StratifiedKFold, KFold, RandomizedSearchCV
from sklearn.ensemble import RandomForestClassifier, ExtraTreesClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
features=ExtraTreesClassifier()
```

```python
features.fit(X,y)
```

```python
imp=pd.Series(features.feature_importances_,index=X.columns) #extracting the important features for the model
```

```python
plt.figure(figsize=(18,6))
imp.nlargest(20).plot.barh()
plt.grid(axis='x')
```

```python
X_new=pd.DataFrame(df[imp.nlargest(20).index])
```

```python
X.shape
```

```python
X_new.shape
```

```python
strata=StratifiedKFold(n_splits=10)
```

```python
for train_idx, test_idx in strata.split(X,y):
    X_train, X_test=X.iloc[train_idx], X.iloc[test_idx]
    y_train, y_test=y.iloc[ train_idx], y.iloc[test_idx]
```

```python
for train_idx, test_idx in strata.split(X_new,y):
    X_new_train, X_new_test=X_new.iloc[train_idx], X_new.iloc[test_idx]
    y_new_train, y_new_test=y.iloc[ train_idx], y.iloc[test_idx]
```

```python
deci=DecisionTreeClassifier()
rand=RandomForestClassifier()
```

```python
n_estimators=[int(i) for i in np.linspace(100,2000,8)]
max_features=['auto','sqrt']
max_depth=[5,15,25]
min_samples_split=[2,10,30,60]
min_samples_leaf=[1,4,9]
```

```python
X_sample=df.iloc[:2000,:-1]
y_sample=df.iloc[:2000,-1]
```

```python
for train_index,test_index in strata.split(X_sample,y_sample):
    X_sample_train,X_sample_test=X_sample.iloc[train_index],X_sample.iloc[test_index]
    y_sample_train,y_sample_test=y_sample.iloc[train_index],y_sample.iloc[test_index]
```

```python
parameters={'n_estimators':n_estimators,
            'max_features':max_features,
            'max_depth':max_depth,
            'min_samples_split':min_samples_split,
            'min_samples_leaf':min_samples_leaf}
parameters
```

```python
r_model=RandomizedSearchCV(rand,
                           param_distributions=parameters,
                           n_jobs=1,
                           cv=5,
                           verbose=2,
                           random_state=42)
```

```python
r_model.fit(X_sample_train,y_sample_train)
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
r_model.best_params_
```

```python
y_pred=r_model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
matrix=confusion_matrix(y_test, y_pred)
```

```python
matrix
```

```python
print(f'Accuracy={np.ceil((matrix[0][0]+matrix[1][1])/matrix.sum()*100)}%\n')
print(f'Error Rate={np.ceil(((matrix[0][1]+matrix[1][0])/matrix.sum())*100)}%\n')
print(f'Precision={np.ceil((matrix[0][0])/(matrix[0][0]+matrix[1][0])*100)}%\n')
print(f'Recall={np.ceil((matrix[0][0])/(matrix[0][0]+matrix[0][1])*100)}%\n')
```
